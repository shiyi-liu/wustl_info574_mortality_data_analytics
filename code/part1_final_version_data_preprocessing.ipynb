```python
import numpy as np
import pandas as pd
```

```python
for year in range(2005,2015): 
    file = r'./data/%d_data.csv' %year
    df = pd.read_csv(file)
    df = df.loc[:,['current_data_year','month_of_death','sex','age_recode_27','education_1989_revision','education_2003_revision','education_reporting_flag','activity_code','race_recode_5']]
    df.to_csv(r'./data/%d_predictors.csv'%year,index=False)
```

```python
file = r'./data/2005_predictors.csv'
df_2005 = pd.read_csv(file)
file = r'./data/2006_predictors.csv'
df_2006 = pd.read_csv(file)
file = r'./data/2007_predictors.csv'
df_2007 = pd.read_csv(file)
file = r'./data/2008_predictors.csv'
df_2008 = pd.read_csv(file)
file = r'./data/2009_predictors.csv'
df_2009 = pd.read_csv(file)
file = r'./data/2010_predictors.csv'
df_2010 = pd.read_csv(file)
file = r'./data/2011_predictors.csv'
df_2011 = pd.read_csv(file)
file = r'./data/2012_predictors.csv'
df_2012 = pd.read_csv(file)
file = r'./data/2013_predictors.csv'
df_2013 = pd.read_csv(file)
file = r'./data/2014_predictors.csv'
df_2014 = pd.read_csv(file)
```

```python
df = pd.concat([df_2005,df_2006,df_2007,df_2008,df_2009,df_2010,df_2011,df_2012,df_2013,df_2014],axis=0,ignore_index=True)
```

```python
df.columns
```

```python
df_sex = df['sex']
df_year = df['current_data_year']
df_month = df['month_of_death']
df_age = df['age_recode_27']-5
df_age[df_age<=1]=1
```

```python
df['education_1989_revision']=df['education_1989_revision'].fillna(value=0)
df['education_2003_revision']=df['education_2003_revision'].fillna(value=0)
from copy import deepcopy
df_edu1 = deepcopy(df['education_1989_revision'])
df_edu2 = deepcopy(df['education_2003_revision'])
```

```python
#primary or less
df_edu1[df_edu1 <= 8] = 0
#high school
df_edu1[(df_edu1 >= 9) & (df_edu1 <= 12)] = 1
df_edu2[(df_edu2 >= 2) & (df_edu1 <= 3)] = 1
#college or higher
df_edu1[(df_edu1 >= 13) & (df_edu1 <= 17)] = 2
df_edu2[(df_edu2 >= 4) & (df_edu2 <= 8)] = 2
#not stated
df_edu1[df_edu1==99] = 3
df_edu2[df_edu2==9] = 3
```

```python
df_edu = df_edu1 + df_edu2
```

```python
df_act=df['activity_code'].fillna(value=10)
df_race = df['race_recode_5']
```

```python
data_full_version = pd.concat([df_sex,df_year,df_month,df_age,df_edu,df_act,df_race],axis = 1,join_axes=[df_sex.index])
data_full_version.columns = ['sex','current_data_year', 'month_of_death', 'age','education','activity','race']
```
