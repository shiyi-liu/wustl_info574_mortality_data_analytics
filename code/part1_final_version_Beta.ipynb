```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
import statsmodels.api as sm
```

```python
file = r'C:\Final Project\Final Data\counter.csv'
df = pd.read_csv(file)
```

```python
df_sex = pd.get_dummies(df['sex'],prefix='sex')
df_year = df['current_data_year'] - 2004
df_month = df['month_of_death']
df_edu = pd.get_dummies(df['education'],prefix = 'edu')
df_edu.columns = ['primary or less','high school','college or higher','edu not stated']
df_age = pd.get_dummies(df['age'],prefix='age')
df_age.columns = ['age_0-4','age_5-9','age_10-14','age_15-19','age_20-24','age_25-29','age_30-34','age_35-39','age_40-44','age_45-49','age_50-54','age_55-59','age_60-64','age_65-69','age_70-74','age_75-79','age_80-84','age_85-89','age_90-94','age_95-99','age_100 and over','age_not_stated']
df_act = pd.get_dummies(df['activity'],prefix='activity')
df_race = pd.get_dummies(df['race'],prefix='race')
```

```python
df_year_cub = df_year ** (0.33)
df_year_sqrt = np.sqrt(df_year)
df_year_2 = (df_year)**2
df_year_3 = (df_year)**3
df_year_4 = (df_year)**4
df_month_cub = df_month ** (0.33)
df_month_sqrt = np.sqrt(df_month)
df_month_2 = (df_month)**2
df_month_3 = (df_month)**3
df_month_4 = (df_month)**4
df_year = pd.concat([df_year,df_year_cub,df_year_sqrt,df_year_2,df_year_3,df_year_4],axis = 1,join_axes=[df_year.index])
df_month = pd.concat([df_month,df_month_cub,df_month_sqrt,df_month_2,df_month_3,df_month_4],axis = 1,join_axes=[df_month.index])
df_year.columns = ['year','year^0.33','year^0.5','year^2', 'year^3', 'year^4']
df_month.columns = ['month','month^0.33','month^0.5','month^2', 'month^3', 'month^4']
```

```python
data = pd.concat([df_sex,df_year,df_month,df_age,df_act,df_race,df_edu,df['ICD'],df['occurence']],axis = 1,join_axes=[df_sex.index])
```

```python
# set the data["ICD"] to the type being analyzed
table = shuffle(data[data['ICD']==20])
X = table.iloc[:,:-2]
y = table['occurence']
X = sm.add_constant(X)
model = sm.GLM(y, X, family=sm.families.Poisson()).fit()
print (model.summary())
```
